```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
#Read the CSV file containing edges
edges_df = pd.read_csv("/content/lastfm_asia_edges.csv")

#Create a graph using NetworkX
G = nx.Graph()
edges_df.head()
```

```python
for _, row in edges_df.iterrows():
    node1 = row['node_1']
    node2 = row['node_2']
    G.add_edge(node1, node2)
```

```python
# Calculate node count
node_count = G.number_of_nodes()

# Calculate edge count
edge_count = G.number_of_edges()

# Calculate average degree
total_degree = sum(dict(G.degree()).values())
average_degree = total_degree / node_count
```

```python
# Calculate the diameter
diameter = nx.diameter(G)

print("Diameter of the graph:", diameter)
```

```python
# Calculate the number of connected components
num_components = nx.number_connected_components(G)

print("Number of connected components:", num_components)
```

```python
# Get a generator of connected components
connected_components = nx.connected_components(G)

# Find the largest connected component
largest_component = max(connected_components, key=len)

# Calculate the size of the largest connected component
size_largest_component = len(largest_component)

print("Size of the largest connected component:", size_largest_component)
```

```python
# Calculate the clustering coefficient
clustering_coefficient = nx.average_clustering(G)

print("Clustering Coefficient:", clustering_coefficient)
```

```python
# Degree Centrality
degree_centrality = nx.degree_centrality(G)

# Betweenness Centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Eigenvector Centrality
eigenvector_centrality = nx.eigenvector_centrality(G)

# Katz Centrality
katz_centrality = nx.katz_centrality(G)

# PageRank Centrality
pagerank_centrality = nx.pagerank(G)

print("Degree Centrality:", degree_centrality)
print("Betweenness Centrality:", betweenness_centrality)
print("Eigenvector Centrality:", eigenvector_centrality)
print("Katz Centrality:", katz_centrality)
print("PageRank Centrality:", pagerank_centrality)
```

```python
# Find all k-cores for random k values between 0 and 20
for k in range(21):  # Iterate over k from 0 to 20
    k_core = nx.k_core(G, k=k)
    if len(k_core) > 0:  # Check if the k-core is non-empty
        print(f"k = {k}:")
        print(k_core.nodes())
```

```python
# Find all cliques
all_cliques = list(nx.find_cliques(G))

# Find the maximum and minimum clique
max_clique = max(all_cliques, key=len)
min_clique = min(all_cliques, key=len)

# Report the number and size of the maximum and minimum clique
num_max_clique = len(max_clique)
num_min_clique = len(min_clique)

print("Number of nodes in the maximum clique:", num_max_clique)
print("Number of nodes in the minimum clique:", num_min_clique)
```
